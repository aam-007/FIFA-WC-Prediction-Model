```python
import pandas as pd

# paths
processed_path = "data/processed"
input_file = f"{processed_path}/fifa_world_cup_matches_1930_2022.csv"
output_file = f"{processed_path}/fifa_wc_clean.csv"

# load master dataset
df = pd.read_csv(input_file)

# standardize team names
team_map = {
    "West Germany": "Germany",
    "USSR": "Russia",
    "Czechoslovakia": "Czech Republic",
    "Yugoslavia": "Serbia",
    "Brasil": "Brazil"
}
df["home_team"] = df["home_team"].replace(team_map)
df["away_team"] = df["away_team"].replace(team_map)

# add match outcome
def outcome(row):
    if row["home_goals"] > row["away_goals"]:
        return "H"
    elif row["home_goals"] < row["away_goals"]:
        return "A"
    else:
        return "D"

df["match_outcome"] = df.apply(outcome, axis=1)

# add goal difference
df["goal_diff"] = df["home_goals"] - df["away_goals"]

# save clean dataset
df.to_csv(output_file, index=False)
print("Clean dataset saved:", output_file)
df.head()
```
